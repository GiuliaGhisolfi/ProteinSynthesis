```python
import pandas as pd
```

```python
sequences = pd.read_csv('data/data.txt', header=None, names=['dna_sequence'])
```

```python
sequences
```

```python
# split in 2 columns 'dna_sequence'
def split(sequence):
    index = sequence.split(' ')[0]
    dna_sequence = sequence.replace(index+' ', '')
    return int(index), dna_sequence

new_data = pd.DataFrame(columns=['dna_sequence'])
for sequence in sequences['dna_sequence']:
    #print(sequence)
    index, dna_sequence = split(sequence)
    new_data = pd.concat([new_data, pd.DataFrame({'dna_sequence': [dna_sequence]}, index=[index])])
new_data
```

```python
new_data.to_csv('data/dna_data.csv')
```
