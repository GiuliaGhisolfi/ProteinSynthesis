```python
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
import pandas as pd
```

```python
HUMAN_GENOMA_DATA_PATH = 'data/human_genoma.csv'
df = pd.read_csv(HUMAN_GENOMA_DATA_PATH, index_col=0)
df.head(2)
```

```python
dna_sequence = 'TGAGYYCN'

from src.nucleotides import NucleotidesSymbolsAllocations
import random

dna_sequence = ''.join([random.choice(NucleotidesSymbolsAllocations[n]) for n in dna_sequence])
dna_sequence = Seq(dna_sequence)

mRNA_sequence = dna_sequence.transcribe()
protein_sequence = mRNA_sequence.translate()

print("DNA Sequence:", dna_sequence)
print("mRNA Sequence:", mRNA_sequence)
print("Protein Sequence:", protein_sequence)
```

```python
dna_sequence = SeqRecord(Seq(df['sequence'].iloc[0]), id=df['ID'].iloc[0])

mRNA_sequence = dna_sequence.seq.transcribe()
protein_sequence = mRNA_sequence.translate()

print("DNA Sequence:", dna_sequence.seq)
print("mRNA Sequence:", mRNA_sequence)
print("Protein Sequence:", protein_sequence)
```
